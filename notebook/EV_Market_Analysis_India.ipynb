```python
import pandas as pd
import numpy as np
state_df = pd.read_csv(r"D:\EV_Market_Analysis_India\data\raw\electric_vehicle_sales_by_state.csv")
maker_df = pd.read_csv(r"D:\EV_Market_Analysis_India\data\raw\electric_vehicle_sales_by_makers.csv")
date_df = pd.read_csv(r"D:\EV_Market_Analysis_India\data\raw\dim_date.csv")
```

```python
state_df['date'] = pd.to_datetime(state_df['date'], format='%d-%b-%y')
maker_df['date'] = pd.to_datetime(maker_df['date'], format='%d-%b-%y')
date_df['date']  = pd.to_datetime(date_df['date'], format='%d-%b-%y')
```

```python
state_df = state_df.merge(date_df, on='date', how='left')
maker_df = maker_df.merge(date_df, on='date', how='left')
```

```python
state_df['penetration_rate'] = (
    state_df['electric_vehicles_sold'] / state_df['total_vehicles_sold']
) * 100
```

```python
state_df.to_csv(r"D:\EV_Market_Analysis_India\data\processed\ev_state_cleaned.csv", index=False)
maker_df.to_csv(r"D:\EV_Market_Analysis_India\data\processed\ev_maker_cleaned.csv", index=False)
```

```python
state_224_df = state_df[state_df['fiscal_year'].isin([2022, 2023, 2024])]
```

```python
state_2024_df = state_224_df[state_224_df['fiscal_year'] == 2024]
state_summary_2024 = (
    state_2024_df
    .groupby('state', as_index=False)
    .agg(
        total_ev_sold=('electric_vehicles_sold', 'sum'),
        total_vehicles_sold=('total_vehicles_sold', 'sum')
    )
)
```

```python
state_summary_2024['penetration_rate'] = (
    state_summary_2024['total_ev_sold'] /
    state_summary_2024['total_vehicles_sold']
) * 100
```

```python
top_5_states = state_summary_2024.sort_values(
    by='penetration_rate', ascending=False
)
top_5_states.head(5)
```

```python
bottom_5_states = state_summary_2024.sort_values(
    by='penetration_rate', ascending=True
)
bottom_5_states.head(5)
```

```python
state_yearly = (
    state_224_df
    .groupby(['state', 'fiscal_year'], as_index=False)
    .agg(
        ev_sold=('electric_vehicles_sold', 'sum'),
        total_sold=('total_vehicles_sold', 'sum')
    )
)
state_yearly['penetration_rate'] = (
    state_yearly['ev_sold'] / state_yearly['total_sold']
) * 100
```

```python
pivot_penetration = state_yearly.pivot(
    index='state',
    columns='fiscal_year',
    values='penetration_rate'
).reset_index()
```

```python
state_ev_yearly = (
    state_df[state_df['fiscal_year'].isin([2022, 2024])]
    .groupby(['state', 'fiscal_year'], as_index=False)
    .agg(ev_sold=('electric_vehicles_sold', 'sum'))
)
```

```python
state_ev_pivot = state_ev_yearly.pivot(
    index='state',
    columns='fiscal_year',
    values='ev_sold'
).reset_index()
```

```python
state_ev_pivot = state_ev_pivot[
    (state_ev_pivot[2022] > 0) & (state_ev_pivot[2024] > 0)
]
```

```python
state_ev_pivot['CAGR'] = (
    (state_ev_pivot[2024] / state_ev_pivot[2022]) ** (1/2) - 1
)
```

```python
top_10_states = state_summary_2024.sort_values(
    by='penetration_rate', ascending=False
).head(10)
```

```python
projection_df = top_10_states.merge(
    state_ev_pivot[['state', 'CAGR']],
    on='state',
    how='left'
)
```

```python
projection_df['projected_ev_sales_2030'] = (
    projection_df['total_ev_sold'] * (1 + projection_df['CAGR']) ** 6
)
projection_df['projected_ev_sales_2030'] = projection_df['projected_ev_sales_2030'].round(0)
```

```python
projection_df[
    ['state', 'total_ev_sold', 'CAGR', 'projected_ev_sales_2030']
]
```

```python
projection_df.to_csv(r"D:\EV_Market_Analysis_India\output\state_ev_sales_projection_2030.csv", index=False)
```

```python
revenue_df = (
    state_df[state_df['fiscal_year'].isin([2022, 2023, 2024])]
    .groupby(['fiscal_year', 'vehicle_category'], as_index=False)
    .agg(ev_units=('electric_vehicles_sold', 'sum'))
)
```

```python
revenue_df['avg_price'] = revenue_df['vehicle_category'].map({
    '2-Wheelers': 100000,
    '4-Wheelers': 1500000
})
```

```python
revenue_df['revenue'] = revenue_df['ev_units'] * revenue_df['avg_price']
revenue_df
```

```python
revenue_pivot = revenue_df.pivot(
    index='vehicle_category',
    columns='fiscal_year',
    values='revenue'
).reset_index()
```

```python
revenue_pivot['growth_22_24_pct'] = (
    (revenue_pivot[2024] - revenue_pivot[2022]) / revenue_pivot[2022]
) * 100
revenue_pivot['growth_23_24_pct'] = (
    (revenue_pivot[2024] - revenue_pivot[2023]) / revenue_pivot[2023]
) * 100
```

```python
revenue_pivot.to_csv(r"D:\EV_Market_Analysis_India\output\ev_revenue_growth_summary.csv", index=False)
```
